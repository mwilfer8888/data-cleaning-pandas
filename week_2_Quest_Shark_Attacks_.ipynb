```python
# shows us what were working with
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"
shark_data =  pd.read_excel(url)  # usecols=["Embarked"]
shark_data.head()
```

```python
# replace all sentence and whitespace from the string date before handling to season
shark_data['Date'] = shark_data['Date'].replace('Before ', '', regex=True)
shark_data['Date'] = shark_data['Date'].replace(' ', '-', regex=True)
# Extract all months ans year from the columns date
shark_data['Month'] = shark_data['Date'].str.extract(r'([a-zA-Z]+)')
shark_data['Year']= shark_data['Date'].str.extract(r'(\d{4})')
# Create a Category for all months
def map_month(month_abbr):
    month_abbr_to_full = {
        'Jan': 'January', 'Feb': 'February', 'Mar': 'March', 'Apr': 'April',
        'May': 'May', 'Jun': 'June', 'Jul': 'July', 'Aug': 'August',
        'Sep': 'September', 'Oct': 'October', 'Nov': 'November', 'Dec': 'December'
    }
    return month_abbr_to_full.get(month_abbr, 'Unknown')
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Fall'
    else:
        return 'Unknown'
shark_data['Month'] = shark_data['Month'].apply(map_month)
# Finish the category seasson for the months
shark_data['Season'] = shark_data['Month'].apply(get_season)
```
